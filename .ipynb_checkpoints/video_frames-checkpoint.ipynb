```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import torch
import os
from tqdm import tqdm
os.getcwd()
```

```python
vidcap = cv2.VideoCapture('/home/mskang/Vlog/hj/image_captioning/BLIP/demo.mp4')
total_frames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))
# iterator
success,image = vidcap.read()

frame_list = list()
for i in tqdm(range(total_frames)):
    if i % 800 == 0:
        frame_list.append(image)     # save frame as JPEG file
        success,image = vidcap.read()
  

print("finish! convert video to frame")
```

```python
from PIL import Image
import requests
import torch
from torchvision import transforms
from torchvision.transforms.functional import InterpolationMode

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def load_demo_image(image_numpy, image_size, device):
    raw_image = Image.fromarray(np.uint8(cv2.cvtColor(image_numpy, cv2.COLOR_BGR2RGB))).convert('RGB')   

    w,h = raw_image.size
    display(raw_image.resize((w//5,h//5)))
    
    transform = transforms.Compose([
        transforms.Resize((image_size,image_size),interpolation=InterpolationMode.BICUBIC),
        transforms.ToTensor(),
        transforms.Normalize((0.48145466, 0.4578275, 0.40821073), (0.26862954, 0.26130258, 0.27577711))
        ]) 
    image = transform(raw_image).unsqueeze(0).to(device)   
    return image
```

```python
from models.blip import blip_decoder

image_size = 200


model_url = 'https://storage.googleapis.com/sfr-vision-language-research/BLIP/models/model_base_capfilt_large.pth'
    
model = blip_decoder(pretrained=model_url, image_size=image_size, vit='base')
model.eval()
model = model.to(device)

with torch.no_grad():
    for frame in frame_list:
        image = load_demo_image(frame, image_size, device)
    # beam search
        caption = model.generate(image, sample=False, num_beams=3, max_length=20, min_length=5) 
        # nucleus sampling
        # caption = model.generate(image, sample=True, top_p=0.9, max_length=20, min_length=5) 
        print('caption: '+caption[0])
```
